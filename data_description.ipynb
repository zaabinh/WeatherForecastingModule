## Data description
### Overview: 
The dataset consists of Vietnam's 63 provinces/cities (before merger of provinces) weather, collected daily from 2021-01-01 to 2025-12-01. This dataset provide regular information of weather for touristic purpose: temperature, wind speed, humidity, weather description
### Source: 
Data is collected from Open Meteo Historical Weather API
https://open-meteo.com/en/docs/historical-weather-api
### Data Statistics:
- 113147 rows, 7 columns, each row include a province/city's weather data in one day
Columns descripton:
- location(object): province / city name
- date(object): date
- weather_code(int64): weather code according to WMO code table 4677, each code stand for a weather condition
- temp, temp_max, temp_min: respectively average temperature, maximum temperature, minimum temperature
- wind_speed: average wind speed
- humidity: relative humidity
- weather description: overall weather condition inferred from weather code
